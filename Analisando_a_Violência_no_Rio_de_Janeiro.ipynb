```
media_roubo = df.roubo_veiculo.mean()
media_furto = df.furto_veiculos.mean()
media_recuperacao = df.recuperacao_veiculos.mean()
print('Média de roubo de veículos: ', media_roubo)
print('Média de furto de veículos: ', media_furto)
print('Média de recuperação de veículos: ', media_recuperacao)
```

A partir das médias extraídas anteriormente, podemos inferir o percentual de veículos recuperados em relação ao total daqueles subtraídos de seus proprietários.

```
total = media_roubo + media_furto
print(round(media_recuperacao/total,2))
```
